```python
import pandas as pd               #for data cleaning and analysis
import numpy as np                #to perform mathematical operation
import re                         #to check whether a specifies a set of strings matches
import seaborn as sns             #to visualize the data
import matplotlib.pyplot as plt   #for data visualization and graphical plotting
from matplotlib import style      #for styling the plot
style.use("ggplot")
from nltk.tokenize import word_tokenize      #to divide strings into lists of substrings
from nltk.stem import WordNetLemmatizer      #to link words with similar meanings to one word. 
from nltk.corpus import stopwords            #to filterout useless data
stop_words = set(stopwords.words("english"))
from wordcloud import WordCloud              #to visualize text data
from sklearn.feature_extraction.text import TfidfVectorizer    #to vectorize a text document
from sklearn.model_selection import train_test_split           #to split the datainto training and testing set
from sklearn.linear_model import LogisticRegression            #to perform Logistic Regression on our model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay   #to evaluate the model performance
```

```python
#loading the data from dataset
tweet_df = pd.read_csv('train.csv')
```

```python
#analysing some data
print(tweet_df["tweet"].iloc[0],"\n")
print(tweet_df["tweet"].iloc[1],"\n")
print(tweet_df["tweet"].iloc[2],"\n")
print(tweet_df["tweet"].iloc[3],"\n")
print(tweet_df["tweet"].iloc[4],"\n")
```

```python
#creating a function for data processing
def data_processing(tweet):
    tweet=tweet.lower()
    tweet=re.sub(r"https\S+|www\S+http\S+",'',tweet,flags=re.MULTILINE)
    tweet=re.sub(r"\@w+|\#",'',tweet)
    tweet=re.sub(r"[^\w\s]",'',tweet)
    tweet=re.sub(r"ð",'',tweet)  #removing punctuations
    tweet_tokens=word_tokenize(tweet)
    filtered_tweets=[w for w in tweet_tokens if not w in stop_words] #removing stopwords
    return " ".join(filtered_tweets)
```

```python
#applying pre-processing to text data
tweet_df.tweet=tweet_df['tweet'].apply(data_processing)
```

```python
#removing duplicate data
tweet_df=tweet_df.drop_duplicates("tweet")
```

```python
lemmatizer=WordNetLemmatizer()
def lemmatizing(data):
    tweet=[lemmatizer.lemmatize(word) for word in data]
    return data
```

```python
#lemmatizing the processed data
tweet_df['tweet']=tweet_df['tweet'].apply(lambda x: lemmatizing(x))
```

```python
#vectorizing the text data using TfidVectorizer and create a bigram lanuage model
vect=TfidfVectorizer(ngram_range=(1,2)).fit(tweet_df['tweet'])
```

```python
feature_names=vect.get_feature_names()
print("Number of features: {}\n", format(len(feature_names)))
print("First 200 features: \n", format(feature_names[:20]))
```

```python
#creating a trigram language model
vect=TfidfVectorizer(ngram_range=(1,3)).fit(tweet_df['tweet'])
```

```python
feature_names=vect.get_feature_names()
print("Number of features: {}\n", format(len(feature_names)))
print("First 200 features: \n", format(feature_names[:20]))
```

```python
#separating the data into x and y to build the model
X=tweet_df['tweet']
Y=tweet_df['label']
X=vect.transform(X) #transforming the x data
```

```python
#spliting the data into training and testing set
x_train, x_test, y_train, y_test=train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
#printing the size of training and testing data
print("Size of x_train", (x_train.shape))
print("Size of y_train", (y_train.shape))
print("Size of x_test", (x_test.shape))
print("Size of x_test", (x_test.shape))
```

```python
#for training the data on logistic regression model
logreg = LogisticRegression()
logreg.fit(x_train, y_train) #loading x_train and y_train data on model
logreg_predict = logreg.predict(x_test) #predicting the value for test data
logreg_acc = accuracy_score(logreg_predict, y_test)
print("Test accuarcy: {:.2f}%".format(logreg_acc*100)) #printing accuracy of the model
```

```python
#printing confusion matrix and classification report
print(confusion_matrix(y_test, logreg_predict))
print("\n")
print(classification_report(y_test, logreg_predict))
```

```python
#for visualizing confusion matrix
style.use('classic')
cm = confusion_matrix(y_test, logreg_predict, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=logreg.classes_)
disp.plot()
```

```python
#for performing hyperparameter tuning
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
param_grid = {'C':[100, 10, 1.0, 0.1, 0.01], 'solver' :['newton-cg', 'lbfgs','liblinear']}
grid = GridSearchCV(LogisticRegression(), param_grid, cv = 5)
grid.fit(x_train, y_train)
print("Best Cross validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
logreg_acc = accuracy_score(y_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100)) #printing model accuracy after applying hyperparamenter tuning
```

```python
print(confusion_matrix(y_test, y_pred))
print("\n")
print(classification_report(y_test, y_pred))
```
